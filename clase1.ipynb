```python
size = 100
results = np.random.randint(2,size = size)

x = np.unique(results)
y = np.zeros(len(x))
for i in range(len(x)):
    y[i] = sum(results == x[i])
y = y/len(results)
plt.scatter(x,y)
```

```python
def aproximar_pmf(results):
    size = len(results)
    x = np.unique(results)
    y = np.zeros(len(x))
    for i in range(len(x)):
        y[i] = sum(results == x[i])
    y = y/size
    # Ploteo
    fig,ax = plt.subplots() 
    ax.scatter(x,y)
    ax.set_ylim(0,1)
    ax.set_title(f"Aproximación de la pmf con {size} samples")
    plt.show()
```

```python
samples = np.random.randint(2,size = 100)
aproximar_pmf(samples)
```
