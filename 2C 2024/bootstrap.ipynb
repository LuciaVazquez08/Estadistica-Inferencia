```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats


# Tomar una muestra tamaño de la columna 'population'
sample_size = 100
sample = np.random.randint(0, 801, size=sample_size)
plt.hist(sample)
plt.show()
```

```python
# Cálculo el promedio (media) muestral
mean = np.mean(sample)

# Cálculo el desvío estándar muestral
std = np.std(sample)

# Cálculo el error estándar del promedio muestral
se_mean = std / np.sqrt(sample_size)

# Cálculo el intervalo de confianza asintótico para la media
alpha = 0.05
low_mean = (mean - stats.norm.ppf(1-alpha/2)* se_mean)
up_mean = (mean + stats.norm.ppf(1-alpha/2)* se_mean)
print('mean:', mean)
print("se(mu) =", se_mean)
print('CI(mu):', [low_mean, up_mean])
```
