```python
import numpy as np
import matplotlib.pyplot as plt
```

EJERCICIO 8


```python
lambda_param = 1/2

n_values = range(1, 3001)

means = []

for n in n_values:
    observations = np.random.exponential(1/lambda_param, n)
    mean = np.mean(observations)
    means.append(mean)

# Graficar n vs x_n
plt.figure(figsize=(10, 6))
plt.plot(n_values, means, label='Media de las observaciones')
plt.axhline(y=1/lambda_param, color='r', linestyle='--', label='Valor esperado (1/)')
plt.xlabel('n')
plt.ylabel('x_n')
plt.title('n vs x_n para una distribucin exponencial con  = 1/2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetro lambda para la distribucin exponencial
lambda_param = 1/2

# Nmero de repeticiones
num_repetitions = 1000

# Almacenar los promedios
means_2 = []

# Generar observaciones y calcular el promedio para cada repeticin
for _ in range(num_repetitions):
    observations = np.random.exponential(1/lambda_param, 2)
    mean = np.mean(observations)
    means_2.append(mean)

# Crear el histograma
plt.figure(figsize=(10, 6))
plt.hist(means_2, bins=30, density=True, alpha=0.6, color='g')
plt.xlabel('Promedio de 2 observaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de promedios de 2 observaciones (1000 repeticiones)')
plt.grid(True)
plt.show()
```

```python
# Almacenar los promedios
means_10 = []

# Generar observaciones y calcular el promedio para cada repeticin
for _ in range(num_repetitions):
    observations = np.random.exponential(1/lambda_param, 10)
    mean = np.mean(observations)
    means_10.append(mean)

# Crear el histograma
plt.figure(figsize=(10, 6))
plt.hist(means_10, bins=30, density=True, alpha=0.6, color='b')
plt.xlabel('Promedio de 10 observaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de promedios de 10 observaciones (1000 repeticiones)')
plt.grid(True)
plt.show()
```

```python
# Almacenar los promedios para n = 100
means_100 = []

# Generar observaciones y calcular el promedio para cada repeticin
for _ in range(num_repetitions):
    observations = np.random.exponential(1/lambda_param, 100)
    mean = np.mean(observations)
    means_100.append(mean)

# Crear el histograma para n = 100
plt.figure(figsize=(10, 6))
plt.hist(means_100, bins=30, density=True, alpha=0.6, color='r')
plt.xlabel('Promedio de 100 observaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de promedios de 100 observaciones (1000 repeticiones)')
plt.grid(True)
plt.show()

# Almacenar los promedios para n = 1000
means_1000 = []

# Generar observaciones y calcular el promedio para cada repeticin
for _ in range(num_repetitions):
    observations = np.random.exponential(1/lambda_param, 1000)
    mean = np.mean(observations)
    means_1000.append(mean)

# Crear el histograma para n = 1000
plt.figure(figsize=(10, 6))
plt.hist(means_1000, bins=30, density=True, alpha=0.6, color='m')
plt.xlabel('Promedio de 1000 observaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de promedios de 1000 observaciones (1000 repeticiones)')
plt.grid(True)
plt.show()

# Encontrar parmetros  y  para la distribucin normal
mu = np.mean(means_1000)
sigma = np.std(means_1000)

# Superponer la PDF de la distribucin normal
x = np.linspace(min(means_1000), max(means_1000), 100)
pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

plt.figure(figsize=(10, 6))
plt.hist(means_1000, bins=30, density=True, alpha=0.6, color='m', label='Histograma')
plt.plot(x, pdf, 'k', linewidth=2, label=f'N({mu:.2f}, {sigma:.2f})')
plt.xlabel('Promedio de 1000 observaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de promedios de 1000 observaciones con PDF superpuesta')
plt.legend()
plt.grid(True)
plt.show()
```
